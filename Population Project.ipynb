### Section 1.4 Calculate the Proportion of the Population within different age ranges 

The below section is to calculate the proportion of age ranges in the population and plotting a bar chart to visualise this. Steps are provided below to explain how the data was manipulated to give the dataframe that was used to plot our values.

1. The variable `age_columns` below holds the columns that only contains the columns of all the age ranges.
2. The variable `total_population` holds the total population of all people in all LSOA using the `age_columns` variable only holds informations from those columns.
3. The variable `proportions` is where we compute the calculation where the sum for each age column is divided by the total population to give the proportion of the population that is in that specifc age range.

```python
age_columns = df.columns[df.columns.str.startswith('Age')]
total_population = df.loc[:,age_columns].sum().sum()
proportions = df.loc[:,age_columns].sum()/total_population
proportions.plot(kind='bar',figsize=(10,6))
plt.xlabel('Age Groups', fontsize = 15)
plt.ylabel('Proportion', fontsize=15)
plt.title('Population Proportion For Age Group', fontsize= 20);
```

### Section 2.1 Create and assign Younger and Older Proportions Columns to the dataframe
In this section, we will add new columns to the dataframe: `younger_proportions`, which represents the proportion of people aged 19 and younger living in an LSOA; `older_proportions`, which represents the proportion of people aged 65 and older living in an LSOA. The steps to achieve this  will be outlined below.

1. We will use the `assign` function to create the columns `younger_proportions` and `older_proportions`. For `younger_proportions`, we will sum the four age ranges less than or equal to 19 and divide that sum by the total number of people living in the LSOA. For `older_proportions`, we will sum the five age ranges greater than or equal to 65 and divide that sum by the total number of people living in the LSOA.


```python
df = df.assign(younger_proportions = (df.Age4Under+ df.Age5to9 + df.Age10to14 + df.Age15to19)/ df.Total)
df = df.assign(older_proportions = (df.Age65to69 + df.Age70to74 + df.Age75to79 + df.Age80to84 + df.Age85Over)/ df.Total)
df.head()
```

### Section 2.2 Create of function to calculate the median age and Calculation of Median Age for all LSO Areas
In this section, we create a function called `findmedian`, which will be used in conjunction with the `assign` and `apply` functions to calculate the median ages. Once this is done, we will produce a dataframe that shows the summary statistics (e.g., mean, median, standard deviation).

1. Create a function outlining the steps to calculate the median age.
1. Use the `assign` and `apply` functions in conjunction with the `findmedian` function to create a new column that stores the median value for each LSOA.
1. Using the `.describe` function on the dataframe, we will obtain the summary statistics for all the values in the columns.

```python
ages = np.arange(0,90,5)
def findmedian(counts):
    cumul_frac = counts[['Age4Under', 'Age5to9', 'Age10to14', 'Age15to19', 'Age20to24',
       'Age25to29', 'Age30to34', 'Age35to39', 'Age40to44', 'Age45to49',
       'Age50to54', 'Age55to59', 'Age60to64', 'Age65to69', 'Age70to74',
       'Age75to79', 'Age80to84', 'Age85Over']]/counts.Total
    cumul_frac = cumul_frac.cumsum()
    index = 0
    while index < len(cumul_frac) and cumul_frac.iloc[index] < 0.5:
        index +=1
    prev_fract = 0
    if index > 0 : 
        prev_fract = cumul_frac.iloc[index - 1] 
    remaining_fract = 0.5 - prev_fract
    fract_increase = cumul_frac.iloc[index] - prev_fract 

    if index == len(ages) - 1:
        median_age = ages[index]
    else:
        median_age = ages[index] + 5 * ( remaining_fract/ fract_increase)
    return median_age
```

```python
df = df.assign(median_age = df.apply(findmedian, axis=1))
```

```python
summary_stats = round(df.describe(),2)
summary_stats
```

## Section 3.1: Create function the classify the LSOA into area types

In this section the aim is to create a function that takes an LSOA as the parameter and classifies the LSOA into an area type returning the Area type. 

1. Classify the LSOA into different area types (e.g `London Borough`, `Metropolitan District` etc.).
1. Create a new column in the dataframe which will be comprised of the values type of area an LSOA.


```python
def areatype(partofcode):
    if partofcode[:3] == 'E06':
        AreaType = 'Unitary Authorities'
    elif partofcode[:3] == 'E07':
        AreaType = 'Non-metropolitan Districts'
    elif partofcode[:3] == 'E08':
        AreaType = 'Metropolitan Districts'
    elif partofcode[:3] == 'E09':
        AreaType = 'London Borough'
    else:
        AreaType = 'Unknown'
    return AreaType
```

```python
df = df.assign(AreaType=df.PartOfCode.apply(areatype))
```

## Section 4.2: Add new column of the extreme values seen in the scatter plot

In this section, we will add a new column to the dataframe based on proportion ranges this is done following the steps below:

1. Create a function to classify the LSOAs that match our criteria and assign them a classification category.
1. Create a new column in the dataframe, which will consist of the values representing the type of area each LSOA falls into.


```python
def extreme_values(data):
    if data['younger_proportions'] >= 0.45 and data['older_proportions'] <= 0.18:
       return 'Young Families' 
    elif data['younger_proportions'] <= 0.1 and data['older_proportions'] >= 0.4:
        return 'Pensioners'
    elif data['younger_proportions'] <= 0.1 and data['older_proportions'] <= 0.1:
        return 'Workers'
    else:
        return  'Normal'
```

```python
df['Classification'] = df.apply(extreme_values, axis=1)
df.head()
```
